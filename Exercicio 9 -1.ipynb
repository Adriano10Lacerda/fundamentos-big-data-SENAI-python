```python
# Importa as Bibliotecas

import pandas as pd        # MANIPULAR E TRATAR DADOS
import matplotlib.pyplot as plt   # GERAR GRAFICOS
import os                  # GERENCIAR O SISTEMA OPERACIONAL

# MAPEAR O CAMINHO E ARQUIVOS

caminhoArquivo = '../arquivos/company_sales_data.csv'

#ATRIBUIR OS DADOS A UM DATAFRAME

dados = pd.read_csv(caminhoArquivo)

#VERIFICAR O CABEÇALHO

dados.head()
```

### Gerar o primeiro gráfico

```python
# Convertendo as colunas em listas

listaMeses = dados['month_number'].to_list()
listaLucro = dados['total_profit'].to_list()

# Criar o gráfico em função dos dados anteriores

plt.plot(listaMeses,      # EIXO X 
         listaLucro,      # EIXO Y
         label = 'LUCRO', # RÔTULO DE CATEGORIA
         c = '#680D',     # COR DA LINHA
         lw = 3,          # ESPESSURA DA LINHA
         ls = '--',       # ESTILO DA LINHA
         marker = 'o'     # MARCADOR
)

# ADICIONAR UM TITULO AO GRÀFICO

plt.title('LUCRO POR MÊS', fontsize = 18)

#  ADICIONAR TITULOS AOS EIXOS

plt.xlabel('MESES', fontsize = 10, c = '#2c2E33', fontweight = 'bold')
plt.ylabel('LUCRO', fontsize = 10, c = '#2c2E33', fontweight = 'bold')

# ADICIONANDO LEGENDA

plt.legend()

# CRIAR UM DIRETORIO PARA SALVAR GRÀFICO

#os.mkdir('../arquivos/graficos-imagens')

plt.savefig('../arquivos/graficos-imagens/grafico1.png')

```

```python
#  CONVERTER COLUNAS EM LISTAS

listaNames = dados['month_number'].to_list() # EIXO 

listaCremeDental = dados['toothpaste'].to_list()

listaLimpadorFacial = dados['facewash'].to_list()

listaCremeFacial = dados['facecream'].to_list()

listaSabonete = dados['bathingsoap'].to_list()

listaShampoo = dados['shampoo'].to_list()

listaHidrante = dados['moisturizer'].to_list()

# PLOTAR OS DADOS NO GRAFICO

plt.plot(
         listaMeses, 
         listaCremeDental,
         label = 'Creme Dental',
         marker = 'o',
         color = '#38FA00'
         )
plt.plot(listaMeses, 
         listaLimpadorFacial,
         label = 'Limpador Facial',
         marker = 'o',
         color = '#2A27F5'
         )
plt.plot(listaMeses,
         listaCremeFacial,
         label = 'Creme Facial',
         marker = 'o',
         color = '#4A430C',
         ls = '--'
         )
plt.plot(listaMeses,
         listaSabonete,
         label = 'Sabonete',
         marker = 'o',
         color = '#CC48F4' 
         )
plt.plot(listaMeses,
         listaShampoo,
         label = 'Meses',
         marker = 'o',
         color = '#CC48F4',
         ls = '--')
plt.plot(listaMeses,
         listaHidrante,
         label = 'Hidratante',
         marker = 'o',
         color = '#F3F582'
         )

# ADCIONAR TÍTULOS

plt.title(' HISTORICO DE VENDAS', fontsize = 10, color = '#575757')
plt.xlabel('MESES', fontsize = 8, color = '#575757')
plt.ylabel('UNIDADES VENDIDAS', fontsize = 8, color = '#575757')

plt.legend()

# SALVAR O GRÁFICO COMO IMAGEM

plt.savefig('../arquivos/graficos-imagens/exercicio03.png')
```

### Exercício 04 - 
Leia os dados de vendas de pasta de dente de cada mês e mostre-os
usando um gráfico de dispersão.

Além disso, adicione uma grade no gráfico. o estilo da linha de grade deve “–“.
O gráfico de dispersão deve ficar assim.

```python
# CONVERTER COLUNAS EM LISTAS

listaMeses = dados['month_number'].to_list()
listaVendasCremeDental = dados['toothpaste'].to_list()

# CRIAR UM GRÁFICO DE DISPERSÃO

plt.scatter(listaMeses, listaVendasCremeDental)

# CRIAR LINHAS DE GRADE AO GRÁFICO

plt.grid(True, linewidth = 0.5, linestyle = '--', c = '#575757')

# ADICIONAR TITULOS AOS EIXOS E AO GRÁFICO

plt.xlabel('Meses')
plt.ylabel('Unidades Vendidas')
plt.title('Vendas de creme dental')

# CRIAR UM GRÁFICO DE COLUNAS

plt.bar(
         listaMeses, 
         listaCremeDental,
         label = 'Creme Dental',
         color = '#38FA00'
         ) 
plt.bar(listaMeses, 
         listaLimpadorFacial,
         label = 'Limpador Facial',
         color = '#2A27F5'
         )
```
