```python
# Importando a biblioteca
import pandas as pd

# Mapear o caminho e o nome do arquivo

caminho = 'C:\\Users\\FIC\Documents\\Adriano Python\\003 - Arquivos\\'
arquivo = 'Imobiliária.xlsx'

# Abrir o documento e inserir em um dataframe

tabela = pd.read_excel(caminho + arquivo, sheet_name= 'Base')

# Visualizar o cabeçalho do dataframe
tabela.head()

```

### Calcule a soma, média, máximo, mínimo e contagem da coluna Preço.

```python
tabela.describe()

# Funções
print(tabela['preco'].sum())# Soma
print(tabela['preco'].mean())# Media
print(tabela['preco'].max())# Maior valor da coluna
print(tabela['preco'].min())# Menor valor da coluna
print(tabela['preco'].count())# Quantidade de valores da coluna
```

### Gere um arquivo em Excel para representar as vendas efetuadas pela filial
### Sudeste em 2010. Sendo que no relatório deve conter apenas vendas de casas
### sem garagem.

```python
# Criar um relatorio para mostrar as vendas da região Sudeste em 2010 de casa sem

df_final = tabela[
    (tabela['filial']== 'Sudeste')&
    (tabela['data']>='2010-01-01')&
    (tabela['data']<='2010-12-31')&
    (tabela['tipo']=='Casa')&
    (tabela['garag']==0)
]

# Salvar em um diretorio especifico
caminho_destino = 'C:\\Users\\FIC\\Documents\\Adriano Python\\filiais\\'
arquivo_final = 'relatorio-vendas-sudeste-2010-sem-garag.xlsx'

df_final.to_excel(caminho_destino + arquivo_final, sheet_name='EXEMPLO', index= False)
```
