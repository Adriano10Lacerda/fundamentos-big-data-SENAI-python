#### Matplotlib - primeriros passos

Matplotlib é a principal biblioteca para geração de graficos em python. A biblioteca possui uma infinidade de possibilidades, sendo possivel criarmos desde graficos simples ate outros mais elaborados.

### IMPORTANDO A BIBLIOTECA & CRIANDO GRÁFICO

```python
# IMPORTANDO AS BIBLIOTECAS
import matplotlib.pyplot as plt

import numpy as np

# Criar os arrays (matrizes) com valores 

x = np.array ([1,2,3,4,5])
y = np.array ([2,4,6,8,10])

plt.plot(x,y)


```

### Modificando o tipo de marcador

x = valores do eixo X

y = valores do eixo Y

marker = tipo de marcador

c = cor da linha

```python
plt.plot(x,y,marker = '*', c = 'red')
```

### Mudando como os pontos são ligados e o tipo de linha

ds = a forma com que os pontos são ligados

ls = line style (tipo de linha)

lw = line width (espessura da linha)

```python
plt.plot (x,y, marker = '*', c = 'red', ds = 'steps', ls = '--')
```

### ALTERANDO AS CORES DOS MARCADORES

```python
plt.plot (x,y, marker = '*', c = 'red', ds = 'steps', ls = '--', mec = 'blue')
```

### Alterar as cores dos marcadores

```python
plt.plot(
    x, y,                   #VALORES
    marker='o',             #MARCADOR
    c='green',              #COR DA LINHA
    ls = '--',              #LINE STYLE
    lw = 2,                 #ESPESSURA DA LINHA
    mec = 'red',            #COR BORDA DO MARCADOR 
    markerfacecolor = 'red' #ou MFC Alterar cor interna do marcador
    )

######## ADICIONANDO OUTROS ELEMENTOS AO GRÁFICO ###########

# Adicionando t´tulo dos eixos

plt.xlabel('VALORES DE X', fontsize = 15)
plt.ylabel('VALORES DE Y', fontsize = 15)

# Adicionando título do grafico

plt.title('TITULO DO GRAFICO', fontsize = 20, c = 'purple')

# Adicionando os rótulos dos eixos

plt.xticks(fontsize = 8, c = 'blue')
plt.yticks(fontsize = 8, c = 'orange')
```

### ADICIONANDO OUTROS ELEMENTOS

```python
plt.plot(x,y)
plt.xlabel('Valores de X', fontsize = 12)
plt.ylabel('Valores de Y', fontsize = 12)
plt.title ('Título do gráfico', fontsize = 15, c='purple')
plt.xticks(fontsize = 8, c = 'blue')
plt.yticks(fontsize = 8, c = 'orange') 
```

### MULTIPLAS CURVAS NO MESMO MECANISMO DE GRÁFICO

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, label = 'SENO')
plt.plot(x, z, label = 'COSSENO')

plt.legend()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, label = 'SENO')
plt.plot(x, z, label = 'COSSENO')
plt.plot(x, np.sqrt(x), label = 'RAIZ QUADRADA')

plt.legend()
```
